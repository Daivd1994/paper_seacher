```python
# -*- coding: UTF-8 -*-
from bs4 import BeautifulSoup
import requests

ccf_ai = "https://dblp.uni-trier.de/db/journals/tip/tip27.html"

# from urllib import request
# response = request.urlopen(ccf_ai)
# html = response.read()
# html = html.decode("utf-8")
# print(html)
req = requests.get(url=ccf_ai)
html = req.text
bf = BeautifulSoup(html)
# bf = BeautifulSoup(html)
texts = bf.find_all('div', class_='data')

for text_tag in texts:
    print(text_tag.find_all('span',class_='title')[0].string)
```
